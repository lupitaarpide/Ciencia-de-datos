# Actividad 4.2: Imputación de Datos

**Descripción:** Practica técnicas de imputación de datos.

**Instrucciones:**
1. Usa un dataset con valores nulos
2. Aplica las siguientes técnicas de imputación:
   * Imputación por media
   * Imputación por mediana
   * Imputación por moda
   * Imputación hacia adelante/atrás
3. Compara los resultados
4. Justifica cuándo usar cada método

## 1. Importar librerías y crear dataset con valores nulos

```python
import pandas as pd
import numpy as np

np.random.seed(42)

data = {
    "edad": [25, np.nan, 35, 40, np.nan, 30, 28, np.nan, 45, 33],
    "salario": [30000, 45000, np.nan, 60000, 55000, np.nan, 40000, 50000, np.nan, 35000],
    "categoria": ["A", "B", np.nan, "A", "C", "B", np.nan, "C", "A", np.nan],
    "score": [85, np.nan, 90, np.nan, 78, 88, np.nan, 92, 70, 95]
}

df = pd.DataFrame(data)
print("Dataset original:")
print(df)
print("\nValores nulos por columna:")
print(df.isnull().sum())
```

## 2. Técnicas de Imputación

### 2.1 Imputación por __Media__

```python
df_media = df.copy()

for col in ["edad", "salario", "score"]:
    df_media[col] = df_media[col].fillna(df_media[col].mean())

print("Imputación por media:")
print(df_media)
print("\nValores nulos restantes:", df_media.isnull().sum().sum())
```

### 2.2 Imputación por __Mediana__

```python
df_mediana = df.copy()

for col in ["edad", "salario", "score"]:
    df_mediana[col] = df_mediana[col].fillna(df_mediana[col].median())

print("Imputación por mediana:")
print(df_mediana)
print("\nValores nulos restantes:", df_mediana.isnull().sum().sum())
```

### 2.3 Imputación por __Moda__

```python
df_moda = df.copy()

for col in df_moda.columns:
    df_moda[col] = df_moda[col].fillna(df_moda[col].mode()[0])

print("Imputación por moda:")
print(df_moda)
print("\nValores nulos restantes:", df_moda.isnull().sum().sum())
```

### 2.4 Imputación hacia __Adelante__ y hacia __Atrás__

```python
df_ffill = df.copy()
df_bfill = df.copy()

df_ffill = df_ffill.ffill()
df_bfill = df_bfill.bfill()

print("Imputación hacia adelante")
print(df_ffill)
print("\nImputación hacia atrás")
print(df_bfill)
```

## 3. Comparación de Resultados

```python
cols_num = ["edad", "salario", "score"]

comparacion = pd.DataFrame({
    "Original (sin nulos)": df[cols_num].mean(),
    "Media": df_media[cols_num].mean(),
    "Mediana": df_mediana[cols_num].mean(),
    "Moda": df_moda[cols_num].mean(),
    "FFill": df_ffill[cols_num].mean(),
    "BFill": df_bfill[cols_num].mean()
}).round(2)

print("Comparación de medias por técnica de imputación:")
print(comparacion)
```

## 4. Justificación: ¿Cuándo usar cada método?

- __Media__: Datos numéricos con distribución simétrica y sin valores atípicos.
- __Mediana__  Datos numéricos con outliers o distribución sesgada.
- __Moda__  Variables categóricas o numéricas discretas.
- __FFill__  Series de tiempo donde el valor anterior es una buena estimación.
- __BFill__  Series de tiempo donde el valor siguiente es más representativo.

**Conclusión:**
- La __media__ es simple pero sensible a valores extremos.
- La __mediana__ es más robusta ante outliers.
- La __moda__ es ideal para variables categóricas.
- __FFill/BFill__ son útiles cuando el orden de los datos tiene significado temporal.
